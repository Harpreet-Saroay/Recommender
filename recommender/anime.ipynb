```python
df.drop(columns=['Score','Japanese name','Episodes','Premiered', 'Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6','Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1'], inplace=True)
df.head()
```

```python
df.drop(columns=[ 'Producers', 'Licensors','Studios','Duration', 'Rating', 'Ranked', 'Popularity','Members', 'Favorites', 'Watching', 'Completed', 'On-Hold', 'Dropped','Plan to Watch'],inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(columns=["English name","Aired"], inplace=True)
df.head()
```

```python
df["tags"] = df["Type"] + " " + df["Source"]
df.head()
```

```python
df.drop(columns=["Type","Source"],inplace=True)
```

```python
df.head(10)
```

```python
df['Name'] = df['Name'].str.replace(r'\:.*?\\', '', regex=True)
```

```python
df['features'] = df['Genres'].fillna('') + ' ' + df['tags'].fillna('')
```

```python
df['features']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['features'])
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# create recommendation function
index = pd.Series(df.index, index=df['Name'].astype(str).str.lower())

def recommend_anime(title, top_n=10):
    title = title.lower()
    if title not in index:
        return "Anime not found."
    
    idx = index[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_n+1]
    
    anime_indices = [i[0] for i in sim_scores]
    return df['Name'].iloc[anime_indices]
```

```python
recommend_anime("Naruto")
```

```python
import pickle
with open('anime_model.pkl', 'wb') as f:
    pickle.dump((df, tfidf, cosine_sim), f)
```
