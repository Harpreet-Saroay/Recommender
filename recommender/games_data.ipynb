```python
game["Combined"] = game['developers'].astype(str) + game['categories'].astype(str) + game['genres'].astype(str) + game['required_age'].astype(str) + game['platforms'].astype(str)
```

```python
game.drop(['publishers','categories','genres','required_age','n_achievements','developers','platforms','is_released','additional_content','total_reviews','total_positive','total_negative','review_score','review_score_desc'],axis=1,inplace=True)
```

```python
game.drop(['positive_percentual','metacritic','is_free','price_initial (USD)'],axis=1,inplace=True)
```

```python
# Create a new feature: 'release_year'
# Convert 'release_date' to datetime if it's not already
game['release_date'] = pd.to_datetime(game['release_date'], errors='coerce')
game['release_year'] = game['release_date'].dt.year
```

```python
game.head( )
```

```python
game.drop(['release_date'],axis=1,inplace= True)
```

```python
game['Combined'] = game['Combined'].fillna('')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# create Tf -idf vectorizer

vectorizer = TfidfVectorizer(stop_words='english')
vectorizer
```

```python
tfidf_matrix = vectorizer.fit_transform(game['Combined'])
tfidf_matrix
```
