```python
import pandas as pd 
import numpy as np
```

```python
sr = pd.read_csv("TMDB_tv.csv")
sr.head()
```

```python
sr.info()
```

```python
sr.dropna(axis=0, inplace=True)
sr.info()
```

```python
sr.head()
```

```python
sr = sr.drop(columns = ['number_of_seasons','number_of_episodes','original_language','vote_count','vote_average','adult','backdrop_path','first_air_date','last_air_date','in_production','original_name','popularity','poster_path','status','languages','networks','origin_country','spoken_languages','production_countries','episode_run_time','homepage','tagline'],axis = 1)
```

```python
sr.head()
```

```python
sr['tags'] = sr.apply(lambda x: ''.join(x[['overview','type','genres','created_by','production_companies']]),axis = 1)
```

```python
sr.info()
```

```python
sr = sr.drop(columns = ['overview','type','genres','created_by','production_companies'],axis = 1)
sr.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(sr['tags'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Function to get recommendations
def get_recommendations(name, cosine_sim=cosine_sim):
    if name not in sr['name'].values:
        return f"No recommendations found for '{name}'. Please check the show name."
    idx = sr[sr['name'] == name].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get the top 10 most similar shows (excluding the input show itself)
    show_indices = [i[0] for i in sim_scores]
    return sr['name'].iloc[show_indices]
```

```python
show_name = 'Breaking Bad'  # Example show name
recommendations = get_recommendations(show_name)
print(recommendations)
```

```python
import pickle
with open('recommendation_model.pkl', 'wb') as file:
    pickle.dump(cosine_sim, file)
```
